```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread(r'spider.png')
assert f is not None

gamma_lst=[0.2,0.8,1.2,2]
for gamma in gamma_lst:
    t=np.array([(p/255)**gamma*255 for p in range(0,256)]).astype(np.uint8)
    g=cv.LUT(f,t)

    fig, ax = plt.subplots()
    ax.plot(t)
    img = cv.cvtColor(g,cv.COLOR_BGR2RGB)
    fig, ax = plt.subplots()
    ax.imshow(img)
    plt.show()


```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread(r'spider.png' )
assert f is not None

t1=np.linspace(0,100,50)
t2=np.linspace(100,255,150)
t3=np.linspace(255,255,56)


t=np.concatenate((t1,t2,t3),axis=0).astype(np.uint8)


fig, ax = plt.subplots()
ax.plot(t)
ax.set_aspect('equal')
print(len(t))
assert len(t) == 256
g = cv.LUT(f,t)

img = cv.cvtColor(f,cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(g,cv.COLOR_BGR2RGB)
fig,ax=plt.subplots(2,1,figsize = (8,8))
ax[0].imshow(img)
ax[1].imshow(img1)
plt.show()


```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread(r'shells.tif',cv.IMREAD_GRAYSCALE)
assert f is not None

hist_f = cv.calcHist([f],[0],None,[256],[0,256])
g=cv.equalizeHist(f)
hist_g = cv.calcHist([g],[0],None,[256],[0,256])
fig,ax = plt.subplots(2,1,figsize = (8,8))
ax[0].plot(hist_f)
ax[1].plot(hist_g)

img = cv.cvtColor(f,cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(g,cv.COLOR_BGR2RGB)
fig,ax=plt.subplots(1,2,figsize = (10,10))
ax[0].imshow(img)
ax[1].imshow(img1)
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('zion_pass.jpg', cv.IMREAD_COLOR)
hsv = cv.cvtColor(f, cv.COLOR_BGR2HSV)
h,s,v = cv.split(hsv)

t=np.array([(p/255)**0.45*255 for p in range(0,256)]).astype(np.uint8)
new_s=cv.LUT(s,t)
new_hsv = cv.merge([h,new_s,v])
img = cv.cvtColor(f, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(new_hsv, cv.COLOR_HSV2RGB)
fig,ax = plt.subplots(2,1,figsize=(10,10))
ax[0].imshow(img)
ax[1].imshow(img2)
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('zion_pass.jpg', cv.IMREAD_COLOR)
hsv = cv.cvtColor(f, cv.COLOR_BGR2HSV)
h,s,v = cv.split(hsv)



#map hue to 0-30 range (red yellow)
new_h = np.mod(h, 30).astype(np.uint8)
new_hsv = cv.merge([new_h,s,v])

img = cv.cvtColor(f, cv.COLOR_BGR2RGB)
img1 = cv.cvtColor(new_hsv, cv.COLOR_HSV2RGB)
fig,ax = plt.subplots(2,1,figsize=(10,10))
ax[0].imshow(img)
ax[1].imshow(img1)
```
